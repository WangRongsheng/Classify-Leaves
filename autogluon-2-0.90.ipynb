```python
from autogluon.vision import ImagePredictor
train_dataset = ImagePredictor.Dataset.from_folder('../../input0/train/')
print(train_dataset)
```

```python
time_limit = 120 * 60 # 10mins
predictor = ImagePredictor()
predictor.fit(train_dataset, time_limit=time_limit)
print('Top-1 val acc: %.3f' % predictor.fit_summary()['valid_acc'])
```

```python
test_dataset = ImagePredictor.Dataset.from_folder('./dataset/')
#image_path = test_dataset.iloc[0]
#result = predictor.predict(image_path)
#print(result)
pred = predictor.predict(test_dataset)
print(pred)
```

```python
import autogluon.core as ag
ag.utils.generate_csv(pred.tolist(), './submission.csv')
```
